```python
from matplotlib import style
style.use('ggplot')
days=[1,2,3,4,5]
temperature=[36,35,37.7,40,38.2]
plt.xlabel('days',fontsize=15)
plt.ylabel('temperature',fontsize=10)
plt.plot(days,temperature,color='r',marker='o',linewidth='12')
```

```python
vij_temp=[40.5,43.5,44.6,39.2,38.9]
plt.plot(days,vij_temp,'m*--',linewidth=5,label='vij')
plt.plot(days,temperature,color='g',marker='o',linewidth=5,label='guntur')
plt.title('gnt-temp vs vij-temp')
plt.xlabel('days')
plt.ylabel('temperature')
plt.legend(loc=4)
plt.grid(color='r',linestyle='-',linewidth=2)
plt.show()
```

### histograms
- represented in bins
- bin represent data in points
- plt.hist(ele,bins=count)

```python
l1=np.random.randint(10,45,5)
l1
```

```python
plt.hist(l1,bins=50)
plt.xlabel("age of the student")
plt.ylabel("number of students")
plt.show()
```

```python
#help(plt.hist)
l2=np.random.randint(1,25,100)
plt.hist(l2,bins=50,rwidth=0.5,orientation='horizontal',color="c")
plt.show
```

```python
val=[i for i in range(10,50,5)]
val
```

```python
plt.hist(12,val,color="m",histtype="bar")
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
plt.figure(figsize=(8,5))
plt.hist([l1,l2],val,color=["m","y"],histtype="bar",rwidth=0.5)
plt.show
plt.title("multi histograms")
```

### stackplot

```python
## used to draw stacked area plot
## displays the complete data for visualisation
## it displays in the form of constituents
## it mostly behaves like pie chart
```

```python
x=[9,3,5,4,8]
y=[10,6,4,5,4]
z=[7,8,9,7,4]
a=np.array([8,4,9,6,8])
plt.stackplot(x,y,z,a)
plt.show()
```

```python
f=[1,2,3,4,5]
s={'y1':[5,6,7,8,9],'y2':[1,6,5,4,3],'y3':[2,4,6,7,9]}
plt.stackplot(f,s.values())
plt.show()
```

```python
plt.stackplot(f,s.values(),labels=s.keys())
plt.legend(loc='upper left')
plt.title("Stack plot")
plt.show()
```

```python
plt.boxplot(free_sulphur)
```
